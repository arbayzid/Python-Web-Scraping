```python

import requests
from bs4 import BeautifulSoup
import pandas as pd

data = []
current_page = 1
proceed = True

while proceed:
    url = f"https://books.toscrape.com/catalogue/page-{current_page}.html"
    proxies = ""
    # proxies={'http': 'http://customer-[your_username]:[your_password]_@pr.oxylabs.io:7777'}
    page = requests.get(url, proxies=proxies)
    soup = BeautifulSoup(page.text, "html.parser")

    if soup.title.text == "404 Not Found":
        proceed = False
    else:
        all_books = soup.find_all("li", class_="col-xs-6 col-sm-4 col-md-3 col-lg-3")

        for book in all_books:
            item = {}
            item['Title'] = book.find("img").attrs["alt"]
            item['Price'] = book.find("p", class_="price_color").text[2:]
            item['Stock'] = book.find("p", class_="instock availability").text.strip()
            item['Link'] = "https://books.toscrape.com/catalogue/" + book.find("a").attrs["href"]
            data.append(item)

        current_page += 1

# Print the total number of scraped pages
print("Total pages scraped:", current_page)

# Create the DataFrame with specified column order
df = pd.DataFrame(data, columns=['Title', 'Price', 'Stock', 'Link'])
df
```

```python

from google.colab import files

# Save the DataFrame to an Excel file and print status
excel_filename = "books.xlsx"
df.to_excel(excel_filename, index=False)
print(f"Data saved to Excel file: {excel_filename} with {len(df)} records.")

# Download the Excel file to local machine
files.download(excel_filename)

# Save the DataFrame to a CSV file and print status
csv_filename = "books.csv"
df.to_csv(csv_filename, index=False)
print(f"Data saved to CSV file: {csv_filename} with {len(df)} records.")

# Download the CSV file to local machine
files.download(csv_filename)
```
